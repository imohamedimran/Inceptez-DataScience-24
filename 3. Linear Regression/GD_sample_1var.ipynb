```python
y = lambda x: 4 * x **2 + 4* x + 5
```

```python
y(0)
```

```python
der = lambda x: 8 * x + 4
```

```python
der(0)
```

```python
learning_rate = 0.01
cur_x = 0 #random chosen value
number_of_iteration = 500
change = 0
```

```python
while number_of_iteration >=0:
    prev_x = cur_x
    cur_x = cur_x - der(cur_x) * learning_rate #update the new_x
    change = abs(prev_x - cur_x)
    print('No of iter:', number_of_iteration)
    print("Value of y in the equation: ", y(cur_x))
    print('Value of x in y:', prev_x)
    print('Slope or Derivative:', der(prev_x))
    print('Change of y in x:', change, '\n')
    number_of_iteration -= 1 #number_of_iteration = number_of_iteration - 1
    if change <0.0001:
        break
```
